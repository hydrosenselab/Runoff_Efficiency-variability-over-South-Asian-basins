# 1. Difference in average duringg 1980-2000, 2000-2020

```python
mdl=1
model=["LIS_MERRA2","LIS_IMD","LIS_CHIRPS","LIS_ERA5"]
ssn=["June","July","August","Monsoon","Winter"]
var=['TotalPrecip_tavg','Qs_tavg','RE']
ls=[]
for vr in var:
    lsm_M = xr.open_mfdataset(f"DATA_p/{model[mdl]}_1980_2021.nc",engine='netcdf4', parallel=True,chunks={'time': 'auto'})
    lsm_M=lsm_M*60*60*24*30 #converting to mm/ month
    lsm_M=lsm_M[['TotalPrecip_tavg','Qs_tavg']]
    RE=lsm_M["Qs_tavg"]/lsm_M["TotalPrecip_tavg"]
    lsm_M['RE']=RE
    #lsm_M_c=lsm_M.where(RE<1)
    
    da=lsm_M[vr]
    
    ssn_m=[[6],[7],[8],[6,7,8,9],[1,2,3]]
    for ss in ssn_m:
        months=da["time.month"]
        ds=da.sel(time=months.isin(ss))
        D=ds.groupby("time.year").mean(dim='time').sel(year=slice(2000, 2021)).mean(dim='year')-ds.groupby("time.year").mean(dim='time').sel(year=slice(1980, 2000)).mean(dim='year')
        ls.append(D)
 
```

```python
plt_tit=[[a, b] for a in ssn for b in var]
fig, axs = plt.subplots(5,3,subplot_kw={'projection': ccrs.PlateCarree()},figsize=(14,14*1.6),constrained_layout=False)
axs=axs.flatten()
for i in range(len(ls)):
    xrr=ls[i].compute()
    Map_plot1(xrr,fig,axs[i],plt_tit[i],'2000/2020 - 1980/2000')

plt.savefig(f"2_Preliminary analysis/ssss.jpeg", bbox_inches='tight')
plt.show()
```

```python
## Single map function
def Map_plot1(xrr1,fig,axs,plt_tit,cbar_tit):

    vmin1 = -max(abs(xrr.quantile(0.1)), abs(xrr.quantile(0.9)))
    vmax1 = -vmin1
    cmap1='RdBu'
    
# Step 3 : Figure axis confiureations   
    fig.subplots_adjust(bottom=None, left=None, top=None, right=None, wspace=0.13, hspace=0.13) 
    [x.set_linewidth(1.5) for x in axs.spines.values()]
    axs.set_extent([67.90, 97,3, 37.50], ccrs.PlateCarree())

    lf=axs.get_position().x0
    bt=axs.get_position().y0
    wd=axs.get_position().width
    hg=axs.get_position().height

# Step 5 : Plotting the data
    basin_map = gpd.read_file('DATA_p/Shapefiles/Indian_basin/India_transboundary.shp')
    #ind_map = gpd.read_file('DATA/Shapefiles/Indian_map/India_country.shp')
    Main_plot=xrr1.plot(vmin=vmin1,vmax=vmax1,ax=axs,cmap=cmap1,add_labels=False,add_colorbar=False) 
    basin_map.plot(ax=axs,color = 'none',edgecolor = 'white',linewidth=.4,linestyle='--')
    
    axs.set_title(f"{plt_tit[1]} ({plt_tit[0]})",fontsize=13,pad=8,fontname='Times New Roman',weight='bold')
    
# Step 4: Colourbar settings   

    
    
    cbar_ax = fig.add_axes([lf+wd*0.51, bt + 0.026, wd-wd*0.6, 0.005])
    cbar=plt.colorbar(Main_plot,cax=cbar_ax, orientation="horizontal",extend='neither', label = "prec") #,label = cbar_tit

    #cbar.ax.tick_params(axis='both', labelsize=13) 

    cbar.set_label(cbar_tit, fontsize=wd*45, labelpad=-wd*150,weight="bold")#position=(0.5, -0.5)
    
```
