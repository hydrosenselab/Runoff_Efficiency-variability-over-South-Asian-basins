# 1.Preprocessing

```python
#1. LOADING IMD PRECIPITATION
year=np.arange(1980,2023,1)
files=sorted(glob.glob('DATA_r/2_IMD/RF25_ind*'+'rfp25.nc'))
imd1 = xr.open_mfdataset(files, combine='by_coords',engine='netcdf4', parallel=True,chunks={'time': 'auto'})

# Making name consistent
imd_new = imd1.rename({'LONGITUDE': 'lon', 'LATITUDE': 'lat', 'TIME': 'time'})
imd = imd_new.transpose(*['time','lat','lon'])
imd_M=imd.resample(time='1M').mean()*30 #Monthly conversion
```

```python
#2. Loading LIS output
lsm_M= xr.open_mfdataset('DATA_p/1_LIS_1980_2021_M.nc')
lsm_M=lsm_M*60*60*24*30 #converting to mm/ month
lsm_M=lsm_M.interp(lat=imd_M.lat,lon=imd_M.lon)

imd_lsm = xr.Dataset({'lsm': lsm_M['TotalPrecip_tavg'], 'imd': imd_M['RAINFALL']})

lsm_monthly  =  imd_lsm.groupby('time.month')
lsm_mean     =  imd_lsm.groupby('time.month').mean(dim='time')
lsm_anom = lsm_monthly-lsm_mean  
lsm_std  = imd_lsm.groupby('time.month').std(dim='time')
lsm_norm = (lsm_anom.groupby('time.month'))/lsm_std

imd_lsm = xr.Dataset({'lsm': lsm_M['TotalPrecip_tavg'], 'imd': imd_M['RAINFALL'], 'imd_n': lsm_norm['imd'],'lsm_n': lsm_norm['lsm']})
```
