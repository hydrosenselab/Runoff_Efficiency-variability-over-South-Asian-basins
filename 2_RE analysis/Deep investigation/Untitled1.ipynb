```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values1 = [23, 30, 15, 35, 29]
values2 = [25, 32, 17, 30, 27]

# Set the positions of the bars on the x-axis
x = np.arange(len(categories))

# Set the width of the bars
width = 0.35

# Creating axes object
fig, ax = plt.subplots()

# Plotting the bars
ax.bar(x - width/2, values1, width, label='Group 1')
ax.bar(x + width/2, values2, width, label='Group 2')

# Adding labels, title, and custom x-axis tick labels
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Multiple Bar Plot')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

# Adding value labels on top of the bars
for i in range(len(categories)):
    ax.text(x[i] - width/2, values1[i] + 0.5, str(values1[i]), ha='center')
    ax.text(x[i] + width/2, values2[i] + 0.5, str(values2[i]), ha='center')

# Show plot
plt.show()
```
