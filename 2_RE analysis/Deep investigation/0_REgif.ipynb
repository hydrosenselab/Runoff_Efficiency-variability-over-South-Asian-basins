# 1.Monthly RE variation

### GIF 1

```python
import xarray as xr
import matplotlib.pyplot as plt
import imageio

# Load the NetCDF file using xarray
ds = RE

# Customize font size and type
font_size = 12
font_type = 'Times New Roman'  # Change this to the desired font
font_weight = 'bold'
```

```python
    
for i in range(len(ds.time)):     
    ds.isel(time=i).plot(vmin=0,vmax=1)
    format_time = pd.to_datetime(ds.time[i].values).strftime('%Y-%m-%d')
    
    #axis formates
    plt.title(f'{format_time}',fontsize=14, fontname=font_type, fontweight=font_weight)
    plt.xlabel('Longitute', fontsize=font_size, fontname=font_type, fontweight=font_weight)
    plt.ylabel('Longitute', fontsize=font_size, fontname=font_type, fontweight=font_weight)
    
    plt.savefig(f'output/1_Trendanalysis/image_bytes/RE_{i}.png')
    plt.clf()
    
# Use imageio to create a GIF from the saved plots
# Use imageio to create a GIF from the saved plots
images = [imageio.imread(f'output/1_Trendanalysis/image_bytes/RE_{i}.png') for i in range(len(ds.time))]
imageio.mimsave('output/1_Trendanalysis/RE_animation_M.gif', images,duration=0.1)  
```

# 2.Annual averaged RE Variation

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.io.shapereader import Reader
from cartopy.feature import ShapelyFeature
import geopandas as gpd

import matplotlib
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.cm import ScalarMappable
matplotlib.rcParams.update(matplotlib.rcParamsDefault)
matplotlib.rcParams.update(matplotlib.rcParamsDefault)
```

```python
RE_yrmean=RE.groupby('time.year').mean(dim='time')
ds=RE_yrmean
```

```python
matplotlib.rcParams['font.family'] = 'Times New Roman' # setting default font

fig = plt.figure(figsize=(7,7))
axes = plt.axes(projection=ccrs.PlateCarree())
axes.set_extent([67.90, 97.59, 5.5, 37.50], ccrs.PlateCarree())

gl = axes.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                linewidth=.2, color='black', alpha=0.1, linestyle='-.')
gl.xlabel_style = {"size" : 10,'family': 'Times New Roman','weight': 'bold'}
gl.ylabel_style = {"size" : 10,'family': 'Times New Roman','weight': 'bold'}
gl.top_labels = False
gl.right_labels = False

# Set a custom x-axis label
axes.text(0.5, -0.1, 'Longitude', transform=axes.transAxes,
        fontsize=14.5, ha='center',weight='bold', fontfamily='Times New Roman')
axes.text(-0.155, 0.5, 'Latitude', transform=axes.transAxes,weight='bold',
        fontsize=14.5, fontfamily='Times New Roman', va='center', rotation='vertical')


#shape_feature = ShapelyFeature(Reader('input/shapefile/India-States.shp').geometries(), ccrs.PlateCarree(), linewidth=1.5, facecolor='none', edgecolor='black')
#axes.add_feature(shape_feature)
axes.add_feature(cfeature.OCEAN, facecolor="white", edgecolor='black')
axes.add_feature(cfeature.LAND, linewidth=1.5, facecolor='white', edgecolor='black')
axes.set_facecolor(None)
axes.add_feature(cfeature.STATES, zorder=1, linewidth=0.5, edgecolor='white', linestyle='-.')

#Colour bar   
cmap='viridis'
sm1 = ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=0, vmax=1))
cbar=plt.colorbar(sm1,orientation="vertical", shrink=0.5, pad=.07, aspect=15, label = 'Runoff Efficiency')
cbar.set_label('Runoff Efficiency', fontsize=14,weight="bold")
cbar.ax.yaxis.labelpad = 13
    

def animate(time):
    #print("hi")
    ds.isel(year=time).plot.contourf(vmin=0,vmax=1 ,ax=axes, cmap='viridis', add_labels=False, add_colorbar=False) 
    axes.add_feature(cfeature.STATES, zorder=1, linewidth=0.4, edgecolor='grey', linestyle='-.')
    format_time = ds.year[time].values #%Y-%m-%d
    plt.title(f'Year : {format_time}',fontsize=14, fontname=font_type, fontweight=font_weight)
    
ani = animation.FuncAnimation(fig, animate, 122, interval=1, blit=False)
ani.save('output/1_Trendanalysis/RE_animation_Yg.gif', writer='pillow', fps=2, savefig_kwargs={"transparent": True, "facecolor": "none"})
```
